# LinLogScale and LinLogAxis Demo

A scale that can toggle between linear and log modes, paired with an axis that shows proper minor ticks for both modes.

```python
import numpy as np
import ipywidgets as widgets
from bqplot import Figure, Lines, PanZoom
from bqplot_linlog import LinLogScale, LinLogAxis
```

## Basic example: toggle between linear and log

Create some sample data and plot it. Use the toggle button to switch the y-axis between linear and log mode.

```python
# Sample data: exponential growth with noise
x = np.linspace(0, 10, 200)
y = np.exp(x / 2) + np.random.normal(0, 0.5, len(x))
y = np.clip(y, 0.1, None)  # keep positive for log mode

scale_x = LinLogScale(mode='linear')
scale_y = LinLogScale(mode='linear')
scales = {"x": scale_x, "y": scale_y}

axis_x = LinLogAxis(scale=scale_x, label="x", side="bottom")
axis_y = LinLogAxis(scale=scale_y, label="y", side="left")

line = Lines(x=x, y=y, scales=scales, colors=["steelblue"])

figure = Figure(
    scales=scales,
    axes=[axis_x, axis_y],
    marks=[line],
    title="LinLog Scale Demo",
)
figure.layout.width = "800px"

# Toggle button for y-axis mode
toggle_y = widgets.ToggleButtons(
    options=["linear", "log"],
    value="linear",
    description="Y scale:",
)
widgets.link((toggle_y, "value"), (scale_y, "mode"))

widgets.VBox([toggle_y, figure])
```

## Both axes toggleable with pan/zoom

```python
# Power-law data across several decades
x2 = np.logspace(-1, 4, 300)
y2 = 2.5 * x2**0.75 + np.random.normal(0, 1, len(x2)) * x2**0.3

scale_x2 = LinLogScale(mode='log', min=0.1, max=1e4)
scale_y2 = LinLogScale(mode='log', min=0.1)
scales2 = {"x": scale_x2, "y": scale_y2}

axis_x2 = LinLogAxis(scale=scale_x2, label="x", side="bottom")
axis_y2 = LinLogAxis(scale=scale_y2, label="y", side="left")

line2 = Lines(x=x2, y=y2, scales=scales2, colors=["crimson"])

figure2 = Figure(
    scales=scales2,
    axes=[axis_x2, axis_y2],
    marks=[line2],
    title="Log-Log Plot (toggleable)",
)
figure2.layout.width = "800px"
figure2.interaction = PanZoom(scales={"x": [scale_x2], "y": [scale_y2]})

toggle_x2 = widgets.ToggleButtons(options=["linear", "log"], value="log", description="X scale:")
toggle_y2 = widgets.ToggleButtons(options=["linear", "log"], value="log", description="Y scale:")
widgets.link((toggle_x2, "value"), (scale_x2, "mode"))
widgets.link((toggle_y2, "value"), (scale_y2, "mode"))

widgets.VBox([widgets.HBox([toggle_x2, toggle_y2]), figure2])
```

## Programmatic mode switching

```python
# You can also switch mode programmatically at any time:
scale_y.mode = 'log'
```
